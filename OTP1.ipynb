```python
import random 
import smtplib
from email.message import EmailMessage 

# Generate a 6-digit OTP
def generate_otp():
    otp = ""
    for i in range(6):
        otp += str(random.randint(0, 9))
    return otp

# Email configuration
from_email = "sender email"
server = smtplib.SMTP('smtp.gmail.com', 587)
server.starttls()
server.login(from_email, "sender email password")

# Get the recipient's email
to_mail = input("Enter E-Mail: ")

# Function to compose and send the email
def send_otp(from_email, to_mail, otp):
    msg = EmailMessage()
    msg['Subject'] = "OTP Verification"
    msg['From'] = from_email
    msg['To'] = to_mail
    msg.set_content("Your OTP is: " + otp)
    server.send_message(msg)
    print("Email Sent Successfully")

# Generate and send the initial OTP
otp = generate_otp()
send_otp(from_email, to_mail, otp)

# OTP verification loop
attempts = 3
while attempts > 0:
    input_otp = input("Enter OTP: ")
    if input_otp == otp:
        print("OTP Verified")
        break
    else:
        print("Incorrect OTP. Please try again")
        attempts -= 1
        if attempts == 0:
            print("OTP verification failed. Please try again later.")
            break
        else:
            print(f"You have {attempts} attempt(s) left.")
            resend = input("Do you want to resend the OTP? (y/n): ")
            if resend.lower() == "y":
                otp = generate_otp()  # Generate a new OTP
                send_otp(from_email, to_mail, otp)
```
